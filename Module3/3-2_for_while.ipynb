| 用法 | 說明 |
|:---:|:---:|
| `range(m)` | 走訪 0 到 m-1 的數字 |
| `range(n, m)` | 走訪 n 到 m-1 的數字  |
| `range(n, m, step)` | 以step為間距，走訪 n 到 m-1 的數字 |

##### for

```python
#crtl + d 選擇相同文字做修改(vscode指令)
#for <變數> in <可迭代物件>:
n = 1
for char in "contents":
    print(char, end="")
    print(n, end=",")
    n += 1
```

```python
# range(m) 走訪 0 到 m-1 的數字
for i in range(5):
    print("helllo world")
```

```python
# range(n, m) 走訪 n 到 m-1 的數字
for i in range(1, 11):
    print(i,end=', ')
```

```python
# range(n, m, step) 以step為間距，走訪 n 到 m-1 的數字
for i in range(1, 11, 3):
    print(i)
```

### 小試身手 for  
請產出 a 開頭的 n 個偶數(含 a 本身)  
Ex:  
a = 5, n = 3  
Output : 6 8 10

a = 4, n = 5  
Output : 4 6 8 10 12

```python
a = 5
n = 3

for i in range(n):
    print(a + 1 + i * 2)
```

```python
a = 3
n = 5

for i in range(n):
    if a % 2 == 0:
        print(a+i*2)
    else:
        print(a+1+i*2)
```

```python
a = 10
n = 10
for i in range(n):
    if a % 2 != 0:
        print(a+1+i*2)
    else:
        print(a+i*2)
```

```python
n = 5
for i in range(1, n + 1):
    for j in range(i-1):
        print(" ",end="")
    for k in range(n - i + 1):
        print("* ",end="")
    print()
```

```python
n = 5
for i in range(1, n-1):
    print(" *",end="")
```

##### while

```python
count = 1
while count <= 5:
    print(count, end=", ")
    count += 1 #一定要給結束條件
print("count已經不小於5 結束迴圈")
print("count為",count)
```
