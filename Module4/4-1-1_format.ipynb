|語法|說明|
|:---:|:---:|
|%s |插入  的值到字串中|
|%d |插入  的值到字串中|
|%f |插入  的值到字串中|
|%% |輸出  (類似跳脫字元)|

```python
# 多組文字
data1 = 'Hello'
data2 = 100

#msg = {%s, %s!, ' %('Hello','World')}
msg = 'Message: %s, %s!' % (data1,data2)
print(msg)
```

```python
# 整數

age = 18
msg = 'I am %d years old.' % (age)
msg = '資料輸出的結果 Age = %d ' %(age)
print(age)
print(msg)
```

```python
# 文字與整數
msg = "%s is %d years old." % ("Jarvis", 18)
print(msg)
```

```python
# 指定寬度 (維持 10 個字元長度，預設向右看齊)
msg = "[%10s]" % "Hello"
print(msg)
```

```python
# 靠左對齊 (維持 10 個字元長度，向左對齊)
msg = "[%-10s]" % "Hello"
print(msg)

```

```python
# 指定浮點數位數
msg = "[%.3f]" % 3.14256
print(msg)
```

```python
msg = "[%.2f%%]" % 3.14256
print(msg)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
msg = "[%.2s]" % "Hello"
print(msg)
```

```python
# 空白補 0
msg = "[%0.6f]" % 3.14
print(msg)
```
