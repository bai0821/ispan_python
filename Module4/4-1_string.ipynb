|string方法| 功能 |
|:---:|:---:|
|`string.replace()` | 替換字串 |
|`string.strip()` | 去除兩側空格 |
|`str.upper()` | 字串小寫變大寫 |
|`str.lower()` | 字串大寫變小寫 |

```python
# 字串變數初始化
name = 'Jarvius'
print(name)
print(name[0])
```

```python
# 替換字串
#方法 也是 function 一種
# 定義在 class 底下的 function 可以稱為 method
'''
string.replace()
將 string 中的 str1 替換成 str2
'''
name = 'alex'
new_name = name.replace('ex', 'len')
print(new_name)
print(name)
```
