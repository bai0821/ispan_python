```python
# 初始化一個 list
ids = [1, 2, 3, 4, 5, 6]
print(ids)
```

```python
# 新增元素在 list 尾端
# list.append()
ids = [1, 2, 3, 4, 5, 6]
ids.append(7)
print(ids)
```

```python
#      0  1  2  3  4  5
ids = [1, 2, 3, 4, 5, 6]
print(ids[3])
```

```python
# 修改索引位置的元素
ids = [1, 2, 3, 4, 5, 6]
ids[3] = 99 #直接指定修改
print(ids)
```

```python
# 刪除索引位置的元素
# list.pop(<index>)
ids = [1, 2, 3, 4, 5, 6]
ids.pop(3)
print(ids)
```

```python
# 在指定索引 1，新增元素 9 ，其餘元素往後移
# list.insert(<index>, <value>)
ids = [1, 2, 3, 4, 5, 6]
ids.insert(1, 9)
print(ids)
```

```python
# 移除指定的值
# 與list.pop不同，remove直接刪除指定數值，pop刪除指定索引
# lst.remove(<value>)
ids = [1, 2, 3, 4, 5, 6]
ids.remove(4)
print(ids)
```

```python
#重複執行會放入多個
#當沒有要用變數可以給"_"
ids = [1, 2, 3, 4, 5, 6]
for _ in range(5):
    ids.insert(1, 9)
print(ids)
```

```python
while 9 in ids:
    ids.remove(9)
print(ids)
```

```python
ids = [1, 2, 3, 4, 5, 6]
for _ in range(5):
    ids.append(10)
print(ids)
# 計算 list 內的某個值 共有幾個
# list.count(<value>)
cnt_10 = ids.count(10)
print(cnt_10)
```

```python
# list 的合併
ids1 = [1, 2, 3, 4]
ids2 = [5, 6, 7]

for i in ids2:
    ids1.append(i)
print(ids1)
```

```python
# list 的合併
# list.extend()
ids1 = [1, 2, 3, 4]
ids2 = [5, 6, 7]
ids1.extend(ids2)
print(ids1)
```

### list 搭配 for 迴圈

```python
# 建立一個 1~10 的 list
num_list = []
for i in range(1, 11):
    num_list.append(i)
print(num_list)
```

```python
# 建立一個 1~50 的 list 只要偶數
num_list = []
for i in range(1, 51):
    if i % 2 == 0:
        num_list.append(i)
print(num_list)
```

```python
#錯誤方法，因為記憶體位置
outside = []
inside = []
for i in range(1, 11):
    inside.append(i)
    print(inside, id(inside))
```

```python
# 建立一個 list 裡面要放遞增的 list
# 例如 [[1],[1, 2],[1, 2, 3]...[1,2,...,10]]

outside = []

for i in range(1, 11):
    inside = []
    #print(f"i: {i}")
    for c in range(1, i+1):
        #print(c"i: {c}")
        inside.append(c)
    print(inside, id(inside))
    outside.append(inside)
print(outside)
```

```python
#另一種解法
outside = []
inside = []
for i in range(1, 11):
    inside.append(i)
    outside.append(inside.copy())
print(outside)
```

```python
# enumerate() 同時取得 index 與 value
ids = [10, 2, 35, 4, 15]
for index, value in enumerate(ids):
    print(f"index: {index},value: {value}")
```

### List comprehension 列表推導式
`new_list = [expression for item in iterable [if condition]]`  
  
`list的名稱 = [表達式 for 元素 in 迭代對象 [if 判斷式(...的條件)]]`

```python
#建立一個1-29的list
num_list = []
for i in range(1, 11):
    num_list.append(i)
print(num_list)
```

```python
# list comprehansion
#要運算的內容寫在表達式↓ (指appened內容)
num_list_1_to_10 = [i for i in range(1, 11)]
print(num_list_1_to_10)
```

```python
#1~10的二次方
data = [i**2 for i in range(1, 11)]
print(data)
```

上面題目運用

```python
outside = []

for i in range(1, 11):
    inside = []
    for c in range(1, i+1):
        inside.append(c)
    print(inside)
    outside.append(inside)
print(outside)
```

```python
outside = []
for i in range(1, 11):
    outside.append([c for c in range(1, i+1)])
    print(outside)
```

```python
outside = [[c for c in range(1, i+1)] for i in range(1, 11)]
print(outside)
```

```python
# 建立一個 1~50 的 list 只要偶數
num_list = []
for i in range(1, 51):
    if i % 2 == 0:
        num_list.append(i)
print(num_list)
```

```python
num_list = [i for i in range(1, 51) if i % 2 == 0]
print(num_list)
```
