```python
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange'] # 錯誤的寫法
    }
}
print(fruit_dict)
```

```python
# 使用 for 迴圈讀取 dict 資料1 
# 只會給 key 值
for key in fruit_dict:
    print(key)
```

```python
# 使用 for 迴圈讀取 dict 資料2
# 取得 dict 的 value
for value in fruit_dict.values():
    print(value)
```

```python
# 使用 for 迴圈讀取 dict 資料3
for key, value in fruit_dict.items():
    #print(value)
    #print(value["color"])
    for color in value["color"]:
        print(color)
```

```python
# 取得 dict 的 value
# print(fruit_dict["kiwi"]) 導致keyError
print(fruit_dict.get("apple"))
print(fruit_dict.get("kiwi")) # 沒有該 key 值回傳 none 不會導致程式錯誤
```

```python
# 資料處理
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}
]
```

```python
color_data= {}
color_data["apple"] = "red"
print(color_data)
```

```python
# 將資料整理成 {'apple': ['red', 'green'], 'grape': ['purple', 'green'], 'kiwi': ['yellow', 'green']}
# 看到list for迴圈起手
color_data = {}
for fruit_color in fruit_color_list:
    print(fruit_color)
    for name, color in fruit_color.items():  #用.itesm拿數值
        color_data[name] = [color]
        # 如果沒有該 key 值就執行...
        if color_data.get(name) == None:
            color_data[name] = [color]

        # 如果有該 key 值就執行...
        else:
            color_data[name].append(color)

print(f"最終的color_data: {color_data}")

```

##### `defaultdict(<type>)`

```python
from collections import defaultdict
# 當訪問的鍵不存在時，可以自動創建鍵和預設值，而不是拋出KeyError。
data = defaultdict(int)

# 以int型態的預設值 0 來初始化
print(data["price"])
print(data)
```
