```python
# 建立 function 3
def greeting(msg, name): #有順序差別  #def get_msg_list()
    str_list = msg.split()
    str_list.append(name)
    return str_list

data = greeting("Hello World!", "Jarvis")
print(data)
# 如果忘記順序 直接指定參數變數名稱，等於上面的動作
data = greeting(name = "Jarvis", msg = "Hello World!")
print(data)
```

```python
# 建立 function 4
# 預設參數值，如果一個想給數值一個不想給，那有數值的要放後面
def greeting(name: str, msg = "Good Morning") -> list[str]:
    str_list = msg.split()
    str_list.append(name)
    return str_list

data = greeting("User")
print(data)
data = greeting("User", "Hello World!")
print(data)
```

```python
# 改寫 Module2 practice
def get_discount(total , discount):
    get_discount = []
    for total in get_discount:
        if total >= 2000:
            n = total // 2000
            total -= discount * n #total -= discount * (total // 2000)
            get_discount.append(total)
        return total
print(get_discount(8200, 600))
```

```python
def get_discount(total, discount):
    if total >= 2000:
        total -= discount * (total//2000)
    return total

data = [
    {
        "total": 6000,
        "discount": 200
    },
    {
        "total": 8000,
        "discount": 200
    }
]

for d in data:
    print(d["total"], d["discount"])
    answer = get_discount(d["total"], d["discount"])
    print(f"折扣後金額: {answer}元")
d1 = get_discount(8000,200)
print(d1)
```
